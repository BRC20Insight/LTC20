```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os

startblock = 2468307

endblock = 2498982

currentblock = startblock

index_file = f"output_{startblock}_{endblock}.txt"

while (currentblock <= endblock):
    
    print('currentblock 1:' + str(currentblock))
    
    try:

        with open(index_file, 'w', encoding='utf-8') as file:
            
            print('startblock :' + str(startblock) + ' endblock : ' + str(endblock))

            for block in range(startblock, endblock + 1):

                print('Block:' + str(block))
                # 请求网页内容

                url = 'http://127.0.0.1/block/' + str(block)  # 将URL替换为实际的网页地址
                block_response = requests.get(url)
                block_html_content = block_response.text

                # 使用BeautifulSoup解析HTML
                block_soup = BeautifulSoup(block_html_content, 'html.parser')

                tx_ul = block_soup.find('ul', class_='monospace')  

                # 查找所有链接
                tx_links = tx_ul.find_all('a') 

                # 存储包含目标文本的链接
                target_links = []

                # 遍历每个tx链接
                for tx_link in tx_links:

                    # 获取链接的URL
                    tx_link_url = tx_link.get('href')

                    # print(link_url)

                    # 跳过不是本站内链接的情况
                    if not tx_link_url.startswith('/'):
                        continue

                    # 跳过指向自身的链接
                    if tx_link_url == '/':
                        continue

                    tx_text = tx_link_url.split('/tx/')[1]

                    # 构建完整的链接
                    full_tx_link_url = urljoin('http://127.0.0.1', tx_link_url)        

                    # 请求链接的内容
                    tx_link_response = requests.get(full_tx_link_url)
                    tx_link_html_content = tx_link_response.text

                    # 使用BeautifulSoup解析链接的内容
                    tx_link_soup = BeautifulSoup(tx_link_html_content, 'html.parser')

                    # 查找包含"Inscription Geneses"的元素
                    inscription_geneses_header = tx_link_soup.find(string='Inscription Geneses')

                    if inscription_geneses_header is not None:

                        #临时打印
                        # print(full_tx_link_url)

                        inscription_container = inscription_geneses_header.find_next('div', class_='thumbnails')

                        inscription_links = inscription_container.find_all('a')

                        for inscription_link in inscription_links:

                            # 获取链接的URL
                            inscription_link_url = inscription_link.get('href')

                            full_inscription_link_url = urljoin('http://127.0.0.1', inscription_link_url)

                            inscription_text = inscription_link_url.split('/inscription/')[1]

                            # print(full_inscription_link_url)

                            inscription_link_response = requests.get(full_inscription_link_url)

                            inscription_link_html_content = inscription_link_response.text

                            inscription_link_soup = BeautifulSoup(inscription_link_html_content, 'html.parser')

                            inscription_dt_ctype_element = inscription_link_soup.find('dt', string='content type')

                            if inscription_dt_ctype_element:

                                inscription_dd_ctype_element = inscription_dt_ctype_element.find_next_sibling('dd')

                                # 如果找到就获取content
                                if inscription_dd_ctype_element and inscription_dd_ctype_element.text == 'text/plain;charset=utf-8':

                                    #初始化
                                    inscription_dd_address = ''

                                    content_link_html_content = ''

                                    #指向address属性，获取文本
                                    inscription_dt_address_element = inscription_link_soup.find('dt', string='address')

                                    if inscription_dt_address_element:

                                        inscription_dd_address_element = inscription_dt_address_element.find_next_sibling('dd', class_='monospace')

                                        if inscription_dd_address_element:

                                            inscription_dd_address = inscription_dd_address_element.text

                                            # print(inscription_dd_address)

                                    #指向并打开content链接，获取文本
                                    inscription_dt_content_element = inscription_link_soup.find('dt', string='content')

                                    if inscription_dt_content_element:

                                        inscription_dd_content_element = inscription_dt_content_element.find_next_sibling('dd')

                                        if inscription_dd_content_element and inscription_dd_content_element.a:

                                            content_link_url = inscription_dd_content_element.a['href']

                                            full_content_link_url = urljoin('http://127.0.0.1', content_link_url)

                                            # print(full_content_link_url)

                                            content_link_response = requests.get(full_content_link_url)

                                            content_link_html_content = content_link_response.text

                                            content_link_html_content = content_link_html_content.replace('\n', ' ')

                                            # print(content_link_html_content)

                                    file.write(f"{block},{tx_text},{inscription_text},{inscription_dd_address},{content_link_html_content}\n")
            
                currentblock += 1
                
                print('currentblock 2:' + str(currentblock))
            
    except Exception as e:
        
        index_file = f"output_{currentblock}_{endblock}.txt"
        
        
        if (startblock != endblock):
            
            os.rename(f"output_{startblock}_{endblock}.txt", f"output_{startblock}_{currentblock}.txt")
            
        else:
            
            # 如果当前文件已存在，（仅限于连续socket接口抛异常的情况）
            if os.path.exists(f"output_{startblock}_{currentblock}.txt"):
                
                os.remove(f"output_{startblock}_{currentblock}.txt")
                
            os.rename(f"output_{startblock}_{endblock}.txt", f"output_{startblock}_{currentblock}.txt")
        
        startblock = currentblock

```
